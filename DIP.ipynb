```python
from tkinter import filedialog
from tkinter import *
import tkinter as tk
from tkinter import ttk
from PIL import ImageTk,Image
import cv2
import numpy as np
import datetime
from PIL import ImageTk, Image, ImageEnhance
from PIL import ImageFilter



windo = Tk()
windo.title("Scrollable App")
windo.configure(background='white')
windo.title("DIP Project group 17")
width  = windo.winfo_screenwidth()
height = windo.winfo_screenheight()
print(width,height)
windo.geometry(f'{width}x{height}')
windo.iconbitmap('./images/app.ico')
windo.resizable(0,0)




w = 400;h = 350
size = (w, h)

def upload_im():
    try:
        global im,resized
        imageFrame = tk.Frame(windo)
        imageFrame.place(x=415, y=60)
        path = filedialog.askopenfilename()
        im = Image.open(path)
        resized = im.resize(size, Image.ANTIALIAS)
        tkimage = ImageTk.PhotoImage(resized)
        display = tk.Label(imageFrame)
        display.imgtk = tkimage
        display.configure(image=tkimage)
        display.grid()
        dn1 = tk.Label(windo, text='Original Image ', width=20, height=1, fg="white", bg="maroon1",
                       font=('times', 22, ' bold '))
        dn1.place(x=444, y=20)
    except:
        del im
        noti = tk.Label(windo, text = 'Please upload an Image File', width=33, height=1, fg="black", bg="dodgerblue2",
                            font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)

def cam_cap():
    try:
        global display4,imageFrame4,cp,dn4,cap
        cap = cv2.VideoCapture(0)
        imageFrame4 = tk.Frame(windo)
        imageFrame4.place(x=415, y=60)
        dn4 = tk.Label(windo, text='Camera Capture', width=20, height=1, fg="white", bg="black",
                       font=('times', 22, ' bold '))
        dn4.place(x=444, y=20)
        display4 = tk.Label(imageFrame4)
        display4.grid()
        cp = tk.Button(windo, text='Capture Image', bg="blue", fg="white", width=20,
                       height=1, font=('times', 22, 'italic bold '),command = cam_click,activebackground = 'yellow')
        cp.place(x=440, y=370)
        def show_frame():
            global  img
            _, frame = cap.read()
            frame = cv2.flip(frame, 1)
            # frame = imutils.resize(frame, width=400,height=280)
            cv2image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA)
            rgb = cv2.cvtColor(cv2image, cv2.COLOR_RGBA2RGB)
            img = Image.fromarray(rgb)
            img = img.resize(size, Image.ANTIALIAS)
            imgtk = ImageTk.PhotoImage(image=img)
            display4.imgtk = imgtk
            display4.configure(image=imgtk)
            display4.after(10, show_frame)
        show_frame()
    except:
        cp.destroy()
        dn4.destroy()
        nti = tk.Label(windo, text = 'Camera is not Opening!!', width=33, height=1, fg="white", bg="red",
                            font=('times', 15, ' bold '))
        nti.place(x=844, y=370)
        windo.after(5000, destroy_widget, nti)

def cam_click():
    global im,resized
    display4.destroy()
    imageFrame4.destroy()
    cv2.destroyAllWindows()
    im = img.copy()
    resized = im.resize(size, Image.ANTIALIAS)
    imageFrame5 = tk.Frame(windo)
    imageFrame5.place(x=415, y=60)
    tkimage5 = ImageTk.PhotoImage(im)
    display5 = tk.Label(imageFrame5)
    display5.imgtk = tkimage5
    display5.configure(image=tkimage5)
    display5.grid()
    windo.after(2000, destroy_widget, cp)
    cam_break()

def cam_break():
    cap.release()
    
def crop_image():
    x=50
    y=50
    width=100
    height=100
    try:
        global op, noti
        op = im
        if op is not None:
            # Crop the specified region from the op image
            cropped_image = op.crop((x, y, x + width, y + height))
            tkimage1 = ImageTk.PhotoImage(cropped_image)
            imageFrame1 = tk.Frame(windo)
            imageFrame1.place(x=845, y=60)
            dn1 = tk.Label(windo, text='Cropped Image', width=20, height=1, fg="white", bg="black",
                           font=('times', 22, 'bold'))
            dn1.place(x=874, y=20)
            display1 = tk.Label(imageFrame1)
            display1.imgtk = tkimage1
            display1.configure(image=tkimage1)
            display1.grid()
            op = cropped_image  # Update op with the cropped image
        else:
            noti = tk.Label(windo, text='Please load an image first', width=33, height=1, fg="black", bg="violet red1",
                            font=('times', 15, 'bold'))
            noti.place(x=844, y=370)
            windo.after(5000, destroy_widget, noti)
    except Exception as e:
        print(e)
        noti = tk.Label(windo, text='Error occurred while cropping', width=33, height=1, fg="black", bg="violet red1",
                        font=('times', 15, 'bold'))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)

def gray_filter():
    try:
        global op,noti
        #Orignal Image
        op = im.convert('L')
        #Resized Image
        gray =resized.convert('L')
        resi = gray.resize(size, Image.ANTIALIAS)
        tkimage1 = ImageTk.PhotoImage(resi)
        imageFrame1 = tk.Frame(windo)
        imageFrame1.place(x=845, y=60)
        dn1 = tk.Label(windo, text='Gray Image ', width=20, height=1, fg="white", bg="black",font=('times', 22, ' bold '))
        dn1.place(x=874, y=20)
        display1 = tk.Label(imageFrame1)
        display1.imgtk = tkimage1
        display1.configure(image=tkimage1)
        display1.grid()
    except Exception as e :
        print(e)
        noti = tk.Label(windo, text = 'Please upload an Image ', width=33, height=1, fg="black", bg="violet red1",
                            font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)

def cartoon_filter():
    try:
        global op,noti
        img1 = cv2.cvtColor(np.asarray(im), cv2.COLOR_RGB2BGR)
        gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        gray1 = cv2.medianBlur(gray1, 5)
        edges1 = cv2.adaptiveThreshold(gray1, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
        color1 = cv2.bilateralFilter(img1, 9, 300, 300)
        cartoon1 = cv2.bitwise_and(color1, color1, mask=edges1)
        op = Image.fromarray(cv2.cvtColor(cartoon1, cv2.COLOR_BGR2RGB)    )
        img = cv2.cvtColor(np.asarray(resized), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        gray = cv2.medianBlur(gray, 5)
        edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
        color = cv2.bilateralFilter(img, 9, 300, 300)
        cartoon = cv2.bitwise_and(color, color, mask=edges)
        pil_image = Image.fromarray(cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)    )
        resi = pil_image.resize(size, Image.ANTIALIAS)
        tkimage2 = ImageTk.PhotoImage(resi)
        imageFrame2 = tk.Frame(windo)
        imageFrame2.place(x=845, y=60)
        dn2 = tk.Label(windo, text='Oil Image ', width=20, height=1, fg="black", bg="deep sky blue",
                       font=('times', 22, ' bold '))
        dn2.place(x=874, y=20)
        display2 = tk.Label(imageFrame2)
        display2.imgtk = tkimage2
        display2.configure(image=tkimage2)
        display2.grid()
    except Exception as e :
        print(e)
        noti = tk.Label(windo, text = 'Please upload an Image ', width=33, height=1, fg="black", bg="gold",
                            font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)

def sketch_filter():
    try:
        global op
        img = cv2.cvtColor(np.asarray(im), cv2.COLOR_RGB2BGR)
        output = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        output = cv2.GaussianBlur(output, (3, 3), 0)
        output = cv2.Laplacian(output, -1, ksize=5)
        output = 255 - output
        ret, output = cv2.threshold(output, 150, 255, cv2.THRESH_BINARY)
        op = Image.fromarray(output)
        resi = op.resize(size, Image.ANTIALIAS)
        tkimage3 = ImageTk.PhotoImage(resi)
        imageFrame3 = tk.Frame(windo)
        imageFrame3.place(x=845, y=60)
        dn3 = tk.Label(windo, text='Sketch Image ', width=20, height=1, fg="black", bg="gray80",
                       font=('times', 22, ' bold '))
        dn3.place(x=874, y=20)
        display3 = tk.Label(imageFrame3)
        display3.imgtk = tkimage3
        display3.configure(image=tkimage3)
        display3.grid()
    except:
        noti = tk.Label(windo, text = 'Please upload an Image ', width=33, height=1, fg="white", bg="black",
                            font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)

def face_eye_det_filter():
    try:
        global op,tkimage4
        face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
        eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
        img = cv2.cvtColor(np.asarray(im), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        for (x, y, w, h) in faces:
            img = cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 4)
            roi_gray = gray[y:y + h, x:x + w]
            roi_color = img[y:y + h, x:x + w]
            eyes = eye_cascade.detectMultiScale(roi_gray)
            for (ex, ey, ew, eh) in eyes:
                cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 4)
            break
        else:
            notip1 = tk.Label(windo, text='Face not found in Image', width=33, height=1,
                             fg="white", bg="midnightblue",
                             font=('times', 15, ' bold '))
            notip1.place(x=844, y=370)
            windo.after(5000, destroy_widget, notip1)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        op = Image.fromarray(img)
        resi = op.resize(size, Image.ANTIALIAS)
        tkimage4 = ImageTk.PhotoImage(resi)
        imageFrame4 = tk.Frame(windo)
        imageFrame4.place(x=845, y=60)
        dn4 = tk.Label(windo, text='Face & Eye Image ', width=20, height=1, fg="white", bg="navy",
                           font=('times', 22, ' bold '))
        dn4.place(x=874, y=20)
        display4 = tk.Label(imageFrame4)
        display4.imgtk = tkimage4
        display4.configure(image=tkimage4)
        display4.grid()
    except Exception as e:
        notip = tk.Label(windo, text = 'Face not found in Image ', width=33, height=1, fg="white", bg="midnightblue",
                            font=('times', 15, ' bold '))
        notip.place(x=844, y=370)
        windo.after(5000, destroy_widget, notip)
        print(e)

def save_img():
    try:
        global noti,dna
        ts = datetime.datetime.now()
        filename = "{}.jpg".format(ts.strftime("%Y-%m-%d_%H-%M-%S"))
        op.save('./Captures/'+filename)
        dna = tk.Label(windo, text=filename+' Captured', width=33, height=1, fg="black", bg="spring green",
                            font=('times', 15, ' bold '))
        dna.place(x=44, y=370)
        windo.after(5000, destroy_widget, dna)
    except Exception as e:
        print(e)
        noti = tk.Label(windo, text='Please upload an Image', width=33, height=1, fg="black", bg="cyan2",
                        font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)
        
        
     
    
def color_balance():
    global op, noti
    try:
        img = cv2.cvtColor(np.asarray(im), cv2.COLOR_RGB2BGR)
        img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
        img_yuv[:, :, 0] = cv2.equalizeHist(img_yuv[:, :, 0])
        img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)
        op = Image.fromarray(cv2.cvtColor(img_output, cv2.COLOR_BGR2RGB))
        resi = op.resize(size, Image.ANTIALIAS)
        tkimage6 = ImageTk.PhotoImage(resi)
        imageFrame6 = tk.Frame(windo)
        imageFrame6.place(x=845, y=60)
        dn6 = tk.Label(windo, text='Color Balanced Image ', width=20, height=1, fg="white", bg="black",
                       font=('times', 22, ' bold '))
        dn6.place(x=874, y=20)
        display6 = tk.Label(imageFrame6)
        display6.imgtk = tkimage6
        display6.configure(image=tkimage6)
        display6.grid()
    except Exception as e:
        print(e)
        noti = tk.Label(windo, text='Please upload an Image', width=33, height=1, fg="black", bg="cyan2",
                        font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)    
    
    
def sharpen_image():
    global op, noti
    op = im
    try:
        sharpened = op.filter(ImageFilter.SHARPEN)
        resi = sharpened.resize(size, Image.ANTIALIAS)
        tkimage8 = ImageTk.PhotoImage(resi)
        imageFrame8 = tk.Frame(windo)
        imageFrame8.place(x=845, y=60)
        dn8 = tk.Label(windo, text='Sharpened Image ', width=20, height=1, fg="white", bg="black",
                       font=('times', 22, ' bold '))
        dn8.place(x=874, y=20)
        display8 = tk.Label(imageFrame8)
        display8.imgtk = tkimage8
        display8.configure(image=tkimage8)
        display8.grid()
    except Exception as e:
        print(e)
        noti = tk.Label(windo, text='Please upload an Image', width=33, height=1, fg="black", bg="cyan2",
                        font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)    
    
def smooth_image():
    try:
        global im, noti
        op = im
        img_cv = cv2.cvtColor(np.asarray(im), cv2.COLOR_RGB2BGR)
        blurred_img = cv2.GaussianBlur(img_cv, (5, 5), 0)
        op = Image.fromarray(cv2.cvtColor(blurred_img, cv2.COLOR_BGR2RGB))
        resi = op.resize(size, Image.ANTIALIAS)
        tkimage = ImageTk.PhotoImage(resi)
        imageFrame = tk.Frame(windo)
        imageFrame.place(x=845, y=60)
        dn = tk.Label(windo, text='Smoothed Image', width=20, height=1, fg="white", bg="black", font=('times', 22, 'bold'))
        dn.place(x=874, y=20)
        display = tk.Label(imageFrame)
        display.imgtk = tkimage
        display.configure(image=tkimage)
        display.grid()
    except Exception as e:
        print(e)
        noti = tk.Label(windo, text='Please upload an Image', width=33, height=1, fg="black", bg="violet red1",
                        font=('times', 15, 'bold'))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)    
        
        
def detect_edges():
    global op, noti
    op = im
    try:
        edge_detected = op.filter(ImageFilter.FIND_EDGES)
        resi = edge_detected.resize(size, Image.ANTIALIAS)
        tkimage9 = ImageTk.PhotoImage(resi)
        imageFrame9 = tk.Frame(windo)
        imageFrame9.place(x=845, y=60)
        dn9 = tk.Label(windo, text='Edge Detected Image ', width=20, height=1, fg="white", bg="black",
                       font=('times', 22, ' bold '))
        dn9.place(x=874, y=20)
        display9 = tk.Label(imageFrame9)
        display9.imgtk = tkimage9
        display9.configure(image=tkimage9)
        display9.grid()
    except Exception as e:
        print(e)
        noti = tk.Label(windo, text='Please upload an Image', width=33, height=1, fg="black", bg="cyan2",
                        font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)        
      
    
def emboss_image():
    global op, noti
    op = im
    try:
        embossed = op.filter(ImageFilter.EMBOSS)
        resi = embossed.resize(size, Image.ANTIALIAS)
        tkimage10 = ImageTk.PhotoImage(resi)
        imageFrame10 = tk.Frame(windo)
        imageFrame10.place(x=845, y=60)
        dn10 = tk.Label(windo, text='Embossed Image ', width=20, height=1, fg="white", bg="black",
                        font=('times', 22, ' bold '))
        dn10.place(x=874, y=20)
        display10 = tk.Label(imageFrame10)
        display10.imgtk = tkimage10
        display10.configure(image=tkimage10)
        display10.grid()
    except Exception as e:
        print(e)
        noti = tk.Label(windo, text='Please upload an Image', width=33, height=1, fg="black", bg="cyan2",
                        font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)    
        
        
def enhance_image():
    global op, noti
    op = im
    try:
        factor = 1.5  
        enhancer = ImageEnhance.Contrast(op)
        op = enhancer.enhance(factor)
        resi = op.resize(size, Image.ANTIALIAS)
        tkimage7 = ImageTk.PhotoImage(resi)
        imageFrame7 = tk.Frame(windo)
        imageFrame7.place(x=845, y=60)
        dn7 = tk.Label(windo, text='Enhanced Image ', width=20, height=1, fg="white", bg="black",
                       font=('times', 22, ' bold '))
        dn7.place(x=874, y=20)
        display7 = tk.Label(imageFrame7)
        display7.imgtk = tkimage7
        display7.configure(image=tkimage7)
        display7.grid()
    except Exception as e:
        print(e)
        noti = tk.Label(windo, text='Please upload an Image', width=33, height=1, fg="black", bg="cyan2",
                        font=('times', 15, ' bold '))
        noti.place(x=844, y=370)
        windo.after(5000, destroy_widget, noti)

        
        
        

def leave():
    from tkinter import messagebox
    if messagebox.askokcancel("Quit", "Do you want to quit?"):
        windo.destroy()
windo.protocol("WM_DELETE_WINDOW", leave)

def destroy_widget(widget):
    widget.destroy()

dn = tk.Label(windo, text='Digital Image Prossesor', width=20, height=1, fg="white", bg="blue2",
              font=('times', 22, ' bold '))
dn.place(x=24, y=20)



sad_img = ImageTk.PhotoImage(Image.open("./images/logo.png"))
panel4 = Label(windo, image=sad_img)
panel4.pack()
panel4.place(x=20, y=60)

up = tk.Button(windo,text = 'Upload Image',bg="spring green", fg="black", width=20,
                   height=1, font=('times', 22, 'italic bold '),command = upload_im, activebackground = 'yellow')
up.place(x=20, y=340)

up1 = tk.Button(windo,text = 'Camera Capture',bg="midnightblue", fg="white", width=20,
                   height=1, font=('times', 22, 'italic bold '),command = cam_cap, activebackground = 'yellow')
up1.place(x=20, y=410)

hat1 = PhotoImage(file = "./images/sketch.png")
hat_f = tk.Button(windo,borderwidth=0,bg = 'white', image = hat1 ,command = sketch_filter)
hat_f.place(x=850, y=480)

sk = tk.Label(windo, text="Sketch", bg="red", fg="black", width=11,
                   height=1, font=('times', 16, 'italic bold '))
sk.place(x=850, y=587)



#crop
new_button1 = PhotoImage(file="./images/crop.png")
crop_x = 50   
crop_y = 50  
crop_width = 100  
crop_height = 100  
button1 = tk.Button(windo, borderwidth=0, bg='white', image=new_button1, command = crop_image)
button1.place(x=50, y=480)
label1 = tk.Label(windo, text="Crop", bg="red", fg="black", width=11, height=1, font=('times', 16, 'italic bold'))
label1.place(x=48, y=587)

sg = PhotoImage(file = "./images/cartoon.png")
sg_f = tk.Button(windo,borderwidth=0,bg = 'white', image = sg,command = cartoon_filter )
sg_f.place(x=450, y=480)

ca = tk.Label(windo, text="oily", bg="red", fg="black", width=11,
                   height=1, font=('times', 16, 'italic bold '))
ca.place(x=448, y=587)

dog_p = PhotoImage(file = "./images/bw.png")
dog_f = tk.Button(windo,borderwidth=0,bg = 'white', image = dog_p ,command = gray_filter)
dog_f.place(x=250, y=480)

bw = tk.Label(windo, text="Gray image", bg="red", fg="black", width=11,
                   height=1, font=('times', 16, 'italic bold '))
bw.place(x=248, y=587)

snap_p = PhotoImage(file = "./images/save.png")
snap = tk.Button(windo,borderwidth=0,bg = 'white', image = snap_p ,command = save_img)
snap.place(x=1350, y=270)

fd_p = PhotoImage(file = "./images/fd.png")
fd = tk.Button(windo,borderwidth=0,bg = 'white', image = fd_p,command = face_eye_det_filter )
fd.place(x=1050, y=630)

f = tk.Label(windo, text="Face Detection", bg="violet red1", fg="black", width=11,
                   height=1, font=('times', 16, 'italic bold '))
f.place(x=1050, y=737)

sv = tk.Label(windo, text="Save Image", bg="red", fg="black", width=11,
                   height=1, font=('times', 16, 'italic bold '))
sv.place(x=1350, y=400)

quit_p = PhotoImage(file = "./images/quit.png")
q = tk.Button(windo, borderwidth=0,bg = 'white', image = quit_p,command = leave )
q.place(x=1350, y=50)







cb_p = PhotoImage(file="./images/color_balance.png")
cb = tk.Button(windo, borderwidth=0, bg='white', image=cb_p, command=color_balance)
cb.place(x=250, y=630)

cb_label = tk.Label(windo, text="Color Balance", bg="red", fg="black", width=11,
                   height=1, font=('times', 16, 'italic bold '))
cb_label.place(x=250, y=737)





new_button2 = PhotoImage(file="./images/smooth.png")

button2 = tk.Button(windo, borderwidth=0, bg='white', image=new_button2, command=smooth_image)
button2.place(x=450, y=630)

label2 = tk.Label(windo, text="Smoothing", bg="red", fg="black", width=11, height=1, font=('times', 16, 'italic bold'))
label2.place(x=450, y=737)


enhance_p = PhotoImage(file="./images/enhance.png")  

enhance_button = tk.Button(windo, borderwidth=0, bg='white', image=enhance_p, command=enhance_image)
enhance_button.place(x=850, y=630)

enhance_label = tk.Label(windo, text="Image Enhancement", bg="red", fg="black", width=16,
                   height=1, font=('times', 16, 'italic bold '))
enhance_label.place(x=848, y=737)


sharp_p = PhotoImage(file="./images/sharpen.png")  

sharp_button = tk.Button(windo, borderwidth=0, bg='white', image=sharp_p, command=sharpen_image)
sharp_button.place(x=650, y=630)

sharp_label = tk.Label(windo, text="Image Sharpening", bg="red", fg="black", width=15,
                   height=1, font=('times', 16, 'italic bold '))
sharp_label.place(x=648, y=737)


edge_p = PhotoImage(file="./images/edge.png")  

edge_button = tk.Button(windo, borderwidth=0, bg='white', image=edge_p, command=detect_edges)
edge_button.place(x=650, y=480)

edge_label = tk.Label(windo, text="Edge Detection", bg="red", fg="black", width=11,
                   height=1, font=('times', 16, 'italic bold '))
edge_label.place(x=648, y=587)



emboss_p = PhotoImage(file="./images/emboss.png")  

emboss_button = tk.Button(windo, borderwidth=0, bg='white', image=emboss_p, command=emboss_image)
emboss_button.place(x=50, y=630)

emboss_label = tk.Label(windo, text="Embossing", bg="red", fg="black", width=11,
                       height=1, font=('times', 16, 'italic bold '))
emboss_label.place(x=50, y=737)


windo.mainloop()
```
